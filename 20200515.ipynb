```r
f <- read.csv("test1.csv",header=TRUE)
```

```r
#visualization
hist(f$x_1) #hist(f[,1])
boxplot(f$x_1)
```

```r
#relationship between two variables
plot(f$x_1,f$x_2) #scatterplot ※scatter 散らす
```

```r
#single regression model
plot(f$x_1,f$x_2)
res=lm(f$x_1~f$x_2)
summary(res)
```

```r
#Statistical indices of one variable
sd(f$x_1) #standard deviation
mean(f$x_1)
#Statistical indices between two variables
cor(f$x_1,f$x_2) #correlation coefficient
var(f$x_1,f$x_2) #co-variance
```
