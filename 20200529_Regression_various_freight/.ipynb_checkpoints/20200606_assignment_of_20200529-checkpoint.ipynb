```r
dt = read.csv("freight4y.csv", header=T)
dt[1:5,]
#dt[nrow(dt),]

z = array(0, c(4,8)) #arrayで行列を生成((4,8)行列，　要素は全て0)
#z

y = array(seq(2000,2015,by=5)) #seq()でベクトルを生成(seq(start, stop, by=step))
#y
```

```r
dt[1:4,"total"]
```

```r
#4年間のtotal weightをグラフ化

#dt[(dt[,1]==2000), 6]のようにするとdtの中からスライシングに一致した値(他にも行，列)を出力してくれる．
for(i in 1:4){
    for(j in 1:8){
        z[i,j] = sum(as.numeric(  dt[(dt[,1]==y[i]), j+5]  )) #as.numeric()で実数に型変換している ※class()で型を調べられる
    }
}

rownames(z) = y
colnames(z) = colnames(dt)[6:13]
z

barplot(t(z))#, legend=T) #t()でdata.frame, matrixなどを転置できる．(numericやdata.frameを転置するとmatrixになる)
```

```r
#変数間の関係をscatter plotで描画

par(ask=T) #グラフィックスパラメータを調整できる

##2015年のpopulationとGRPの関係を描画
y2015 = dt[dt[,1]==2015,]
plot(log10(y2015[,4]),log10(y2015[,5]),type="n")
text(log10(y2015[,4]),log10(y2015[,5]),dt[1:47,"pref"]) #text(x, y, labels)でグラフ上の点にラベルが描かれる．最後のラベルに指定しているのはdt[1:47,3]でも良い

##各年のpopulationとtotalの関係を描画
for(i in 1:4){
    a = dt[dt[,1]==y[i],]
    plot(log10(a[,4]),log10(a[,14]),type="n")
    text(log10(a[,4]),log10(a[,14]),dt[1:47,"pref"]) #最後のラベルに指定しているのはdt[1:47,3]でも良い
}
```

```r
#agricultureをpopulationとGRPから予測
res.1 <- lm( log(dt[,6]) ~ log(dt[,4])+log(dt[,5]) )
summary(res.1)

#woodをpopulationとGRPから予測
res.2 <- lm( log(dt[,7]) ~ log(dt[,4])+log(dt[,5]) )
summary(res.2)
```

```r
# 全ての列を用いて予測1
#各年，各県のtotal以外(agriculture　wood　mine　machine　chemical　Smachine　miscind　special)を各年のpopulation, GRPから予測
all <- array(0,c(47*4*8,17))
ns <- 0
for(i in 1:8){
    for(j in 1:(47*4)){
        ns <- ns+1
        all[ns,1] <- log(dt[j,i+5])
        all[ns,i*2 ] <- log(dt[j,4]) #d[,4]はpopulation
        all[ns,i*2+1] <- log(dt[j,5]) #d[,5]はGRP
    }
}
#all[10:15,]
#all[1,2]==all[1+47*4,4] #同じ
all <- data.frame(all)
#all[200:205,]
#all[1:5,"X1"]
res.all <- lm( X1 ~., data=all)
#predict(res.all)
#formula(res.all)
summary(res.all)
```

```r
# 全ての列を用いて予測2
#各年，各県のtotal以外(agriculture　wood　mine　machine　chemical　Smachine　miscind　special)を各年のpopulation, GRP/populationから予測
all <- array(0,c(47*4*8,17))
ns <- 0
for(i in 1:8){
    for(j in 1:(47*4)){
        ns <- ns+1
        all[ns,1] <- log(dt[j,i+5])
        all[ns,i*2 ] <- log(dt[j,4])
        all[ns,i*2+1] <- log(dt[j,5]/dt[j,4])
    }
}
all <- data.frame(all)
res.al2 <- lm( X1 ~., data=all)
formula(res.al2)
summary(res.al2)
```

```r
# 予測したいのは将来のfreight volume
##需要の増加を考慮する
##県ごとの増加/減少率を考慮


#1.populationとGRPからtotalを予測する(強い)モデルを作る
all <- array(0,c(47*4*8,3))
ns <- 0
for(j in 1:(47*4)){
    ns <- ns+1
    all[ns,1] <- log(dt[j,14])
    all[ns,2 ] <- log(dt[j,4])
    all[ns,3] <- log(dt[j,5])
}
all <- data.frame(all)
colnames(all) = c("total",　"Population", "GRP")
#all[1:5,]
res.al2 <- lm( total~., data=all)
#summary(res.al2)

#2.predict()と上記のモデルを使って予測
##2.1 予測するのに使うGRPを予測
###2.1.1 PopulationからGRPを予測するモデルを作成
grp = array(0,c(47*4*8,2))
ns <- 0
for(i in 1:8){
    for(j in 1:(47*4)){
        ns <- ns+1
        grp[ns,1] <- log(dt[j,5]) #log(dt[j,14])
        grp[ns,2 ] <- log(dt[j,4])
        #GRP[ns,i*2+1] <- log(dt[j,5]/dt[j,4])
    }
}
colnames(grp) = c("GRP","Population")
grp.dataframe = data.frame(grp)
#grp.dataframe[200:205,]
res.grp = lm(GRP~., data=grp.dataframe)
#summary(res.grp)

###2.1.2 未来の推定人口からGRPを予測
pop = read.csv(file="futPOP2045.csv",header=T,row.names=1)
t = apply(pop,2,sum)
t = lapply(t, log)
t.dataframe = t(data.frame(t))
colnames(t.dataframe) = c("Population")
t.dataframe = data.frame(t.dataframe) #2行前に転置してしまったので(データフレームは転置すると行列になってしまう)，再びデータフレームにする
pred.grp = predict(res.grp, newdata=t.dataframe)
pred.grp = lapply(pred.grp, exp)
#pred.grp###############

##2.2 totalを予測
pred.grp.dataframe = data.frame(t(data.frame(pred.grp))) #データフレームオブジェクトに変換
#pred.grp.dataframe###############
preddata = cbind(t.dataframe, pred.grp.dataframe) #データフレームを横に結合
colnames(preddata) = c("Population","GRP")
#preddata###############
pred.total = predict(res.al2, newdata=preddata)
pred.total###############

#pred.total.exp = lapply(pred.total, exp)
#pred.total.exp###############
```
